```python
summary = df.drop(['Class'], axis=1)
summary['Average'] = summary.mean(numeric_only=True, axis=1)
summary
```

```python
df = df.sample(frac=1)
df
```

```python
Y = df[['Class']]
X = df.drop(['Class'], axis=1)
```

```python
import matplotlib.pyplot as plt
X.plot()
```

```python
summary['Class'].value_counts()
pd.crosstab(summary['Average'],columns='count',normalize=True)
sns.countplot(x= 'Class',data=summary)
sns.boxplot(x= 'Class',y='Average',data=summary)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_min = mms.fit_transform(X_train)
X_test_min = mms.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
independent_scaler = StandardScaler()
X_train_norm = independent_scaler.fit_transform(X_train)
X_test_norm = independent_scaler.transform(X_test)
```

# Applying Standard Scaler To KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train.values.ravel())
Y_pred = knn.predict(X_test)
print(metrics.accuracy_score(Y_test, Y_pred))
```

```python
target_names = ['Normal','TB']
print(classification_report(Y_test, Y_pred.reshape(-1,1), target_names=target_names))
```

```python
cm = confusion_matrix(Y_test, Y_pred.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='OrRd_r')
```
